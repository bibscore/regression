```python
seed = 50
np.random.seed(seed)
```

```python
df = pd.read_csv('AB2_formation_energy_materials_spacegroup.csv')
df.drop('Unnamed: 0', axis='columns', inplace=True)
df
```

```python
df_clean = df[['hform']].dropna().reset_index(drop=True)
df_clean.plot(c="crimson")
```

# Space Group

```python
df['Space Group'].head()
```

```python
df['Space Group'], unique_categories = pd.factorize(df['Space Group'])
df['Space Group'] = df['Space Group'] + 1
```

# hform Histogram

```python
hform = df['hform'].copy()
plt.hist(hform, bins=8, density=False, label='Data', rwidth=0.9, color="crimson")
```

# Train and test dataset splitting

```python
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set = train_test_split(df, test_size=0.2, shuffle=True, random_state=seed)
```

```python
X = train_set.drop(['hform', 'Material'], axis='columns')
y = train_set['hform']
```

```python
X_test = test_set.drop(['hform', 'Material'], axis='columns')
y_test = test_set['hform']
```

# Ridge Linear Model

```python
from sklearn import metrics
from sklearn import linear_model
```

```python
ridge_model = linear_model.Ridge(alpha=7)
ridge_model.fit(X, y)
```

```python
ridge_pred_train = ridge_model.predict(X)
train_error = metrics.mean_squared_error(y, ridge_pred_train, squared=True)


ridge_pred_test = ridge_model.predict(X_test)
test_error = metrics.mean_squared_error(y_test, ridge_pred_test, squared=True)
```

```python
train_error, test_error
```

```python
plt.plot(range(len(y)), y, label="raw data", c="darkslateblue")
plt.plot(range(len(ridge_pred_train)), ridge_pred_train, c="slateblue", label="train data")
plt.plot(range(len(ridge_pred_test)), ridge_pred_test, c="crimson", label="test data")
plt.legend()
plt.grid()
plt.show()
```

# Lasso Linear Model

```python
lasso_model = linear_model.Lasso(alpha=0.1)
lasso_model.fit(X, y)
```

```python
lasso_pred_train = lasso_model.predict(X)
train_error = metrics.mean_squared_error(y, lasso_pred_train, squared=True)


lasso_pred_test = lasso_model.predict(X_test)
test_error = metrics.mean_squared_error(y_test, lasso_pred_test, squared=True)
```

```python
train_error, test_error
```

```python
plt.plot(range(len(y)), y, label="raw data", c="darkslateblue")
plt.plot(range(len(lasso_pred_train)), lasso_pred_train, c="slateblue", label="train data")
plt.plot(range(len(lasso_pred_test)), lasso_pred_test, c="crimson", label="test data")
plt.legend()
plt.grid()
plt.show()
```

# Overfitting?

Training vs Testing Comparison

```python
ridge_train = metrics.mean_squared_error(y, ridge_pred_train) 
lasso_train = metrics.mean_squared_error(y, lasso_pred_train) 
```

```python
ridge_test = metrics.mean_squared_error(y_test, ridge_pred_test) 
lasso_test = metrics.mean_squared_error(y_test, lasso_pred_test)
```

```python
print("TRAIN: ridge reg = ", ridge_train, "--- lasso reg =", lasso_train)
print("TEST:  ridge reg = ", ridge_test, " --- lasso reg =", lasso_test)
```

Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
num_folds = 5
ridge_scores = cross_val_score(ridge_model, X, y, cv=num_folds, scoring='neg_mean_squared_error')
lasso_scores = cross_val_score(lasso_model, X, y, cv=num_folds, scoring='neg_mean_squared_error')
```

```python
print(f"rigde_scores: {ridge_scores}")
print(f"ridge_scores mean: {np.mean(ridge_scores)}")
print(f"ridge_scores std: {np.std(ridge_scores)}")
print(f"ridge_scores var: {np.var(ridge_scores)}")
```

```python
print(f"lasso_scores: {lasso_scores}")
print(f"lasso_scores mean: {np.mean(lasso_scores)}")
print(f"lasso_scores std: {np.std(lasso_scores)}")
print(f"lasso_scores var: {np.var(lasso_scores)}")
```
